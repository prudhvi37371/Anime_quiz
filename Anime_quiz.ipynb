```python
import pandas as pd
import numpy as np
```

```python
am=pd.read_csv('anime-dataset-2023.csv')
```

```python
am.info()
```

```python
am['Score'] = am['Score'].replace('UNKNOWN', np.nan)
convert_dict = {'Score': float
                }

am = am.astype(convert_dict)
```

```python
am["Aired"] = am["Aired"].str.split(" to ", expand=True)[0] # Split the 'Aired' column on " to " and takes the first element.
am["Aired"] = pd.to_datetime(am["Aired"], format='%b %Y', errors='coerce') # Convert the 'Aired' column to datetime objects with specified format and handle errors.
am["Year"] = am["Aired"].dt.year
```

1q) Which anime has the longest synopsis?

```python
am.loc[am["a1"].idxmax()]
```

2q) What is the average score of anime produced by 'Studio Ghibli'?  
*

```python
count=am.groupby("Studios")
s1=count.get_group("Studio Ghibli")
s1['Score'].mean()
```

3q) How many unique anime titles start with the letter 'A'?
*

```python
n1=am.groupby("Name")
n1.filter(lambda x: x['Name'].str.startswith('A').any()).nunique()
```

4q) Which year has the highest number of anime releases?  


```python
am['Premiered'] = am.Premiered.apply(lambda x: np.nan if x == 'UNKNOWN' else x)
am['Year_of_Release'] = am.Premiered.str.split(' ').str.get(1)
am['Year_of_Release'] = am.to_datetime(am.Year_of_Release)
most_releases_year = am['Year_of_Release'].dt.year.value_counts().idxmax()
print(most_releases_year)
```

5q) Which genre has the highest average number of episodes?

```python
am["Genres"].value_counts().sort_values(ascending=True).tail(50)
```

6q) Which anime has the highest number of members and is rated 'R-17+'?

```python
a = am.groupby('Rating')
# Check for potential issues with the group key
print(am['Rating'].unique()) # Print unique values in 'Rating' column to check for inconsistencies.

# Ensure the key exists and use the correct key
a.get_group(am.str.contains("R-17+"))["Members"].max() # Use the correct group key based on the output of the previous line.
```
